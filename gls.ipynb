```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data = pd.DataFrame()
vif_data["feature"] = amiya.columns
vif_data["VIF"] = [variance_inflation_factor(amiya.values, i)
                          for i in range(len(amiya.columns))]

print(vif_data)
```

```python
amiya=dat[["pengeluaran","penduduk"]]
amiya = sm.add_constant(amiya)
y=dat[["miskin"]]
olsmod=sm.OLS(y,amiya).fit()
residuals = olsmod.resid
cov_matrix = np.diag(residuals**2)
gls_model = sm.GLS(dat.miskin,amiya, sigma=cov_matrix).fit()
print(gls_model.summary())
```
